```python
def fetch_data_for_date(date):
    formatted_date = date.strftime('%d-%m-%Y')
    EOD_data = cm.bhav_copy_equities(formatted_date)
    filtered_data_df = EOD_data[EOD_data['SYMBOL'].isin(nifty_500_list['Symbol'])]
    final_data = filtered_data_df[filtered_data_df['SERIES'] == 'EQ']
    final_data.reset_index(drop=True, inplace=True)
    return pd.DataFrame(final_data)

def filter_data_for_common_symbols(df_list):
    common_symbols = set(df_list[0]['SYMBOL'])  # Start with symbols from the first DataFrame
    
    # Find common symbols among all DataFrames
    for df in df_list[1:]:
        common_symbols = common_symbols.intersection(set(df['SYMBOL']))
    
    # Filter each DataFrame to include only common symbols
    filtered_dfs = [df[df['SYMBOL'].isin(common_symbols)] for df in df_list]
    
    return filtered_dfs

# Get today's date
today = datetime(2024, 4, 15)

# Fetch data for today and the previous three days
df = fetch_data_for_date(today)
df1 = fetch_data_for_date(today - timedelta(days=3))
df2 = fetch_data_for_date(today - timedelta(days=5))
df3 = fetch_data_for_date(today - timedelta(days=6))

# Filter data to include only common symbols
df, df1, df2, df3 = filter_data_for_common_symbols([df, df1, df2, df3])

# Display the filtered DataFrames
print("Today's Data:")
print(df)
print("\nYesterday's Data:")
print(df1)
print("\nDay Before Yesterday's Data:")
print(df2)
print("\nThree Days Ago Data:")
print(df3)
```

```python
def detect_gravestone_doji(df):
    gravestone_doji_stocks = []

    for index, row in df.iterrows():
        if abs(row['OPEN']-row['CLOSE'])<0.05*(row['HIGH']-row['LOW']) and 0.05*(row['HIGH']-max(row['OPEN'],row['CLOSE']))> (max(row['OPEN'],row['CLOSE'])-row['LOW']) and abs(min(row['OPEN'],row['CLOSE'])-row['LOW'])<0.05*(row['HIGH']-row['LOW']):
            gravestone_doji_stocks.append(row['SYMBOL'])
    return gravestone_doji_stocks

# Detect Gravestone Doji candlesticks for stocks
gravestone_doji_stocks = detect_gravestone_doji(df)

print("Stocks in which Gravestone Doji candlesticks were formed:")
for stock in gravestone_doji_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which Gravestone Doji candlesticks were formed': gravestone_doji_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_gravestone_doji.xlsx", index=False)

def detect_hangingman(df, df1, df2,df3):
    hangingman_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for Hanging Man pattern
        if (row1_prev['CLOSE'] > row1_prev['OPEN']) and \
           (row2_prev['CLOSE'] > row2_prev['OPEN']) and \
           (row3_prev['CLOSE'] > row3_prev['OPEN']) and \
           (row1_prev['CLOSE'] > row3_prev['CLOSE']) and \
           (2 * abs(row['OPEN'] - row['CLOSE']) < (min(row['OPEN'], row['CLOSE']) - row['LOW'])) and \
           (row['HIGH'] - max(row['OPEN'], row['CLOSE']) < 0.05 * (row['HIGH'] - row['LOW'])):
                hangingman_stocks.append(symbol)

    return hangingman_stocks

# Call the function with your three dataframes
hangingman_stocks = detect_hangingman(df, df1, df2,df3)

print("Stocks in which hanging man candlesticks were formed:")
for stock in hangingman_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which hangingman candlesticks were formed': hangingman_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_hangingman.xlsx", index=False)

def detect_bearish_engulfing(df, df1, df2,df3):
    bearish_engulfing_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for bullish engulfing pattern
        if (row1_prev['CLOSE'] > row1_prev['OPEN']) and \
           (row2_prev['CLOSE'] > row2_prev['OPEN']) and \
           (row3_prev['CLOSE'] > row3_prev['OPEN']) and \
           (row1_prev['CLOSE'] > row3_prev['CLOSE']) and \
           (row['OPEN'] > row1_prev['CLOSE']) and \
           (row['CLOSE'] < row1_prev['OPEN']):
                bearish_engulfing_stocks.append(symbol)

    return bearish_engulfing_stocks

# Call the function with your three dataframes
bearish_engulfing_stocks = detect_bearish_engulfing(df, df1, df2,df3)

print("Stocks in which bearish engulfing candlesticks were formed:")
for stock in bearish_engulfing_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which bearish engulfing candlesticks were formed': bearish_engulfing_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_bearishengulfing.xlsx", index=False)

def detect_darkcloud(df, df1, df2,df3):
    darkcloud_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for bullish engulfing pattern
        if (row1_prev['CLOSE'] > row1_prev['OPEN']) and \
           (row['CLOSE'] < row['OPEN']) and \
           (row['OPEN'] > row1_prev['HIGH']) and \
           ((abs(row1_prev['OPEN']+row1_prev['CLOSE']))/2 > row['CLOSE']):
                darkcloud_stocks.append(symbol)

    return darkcloud_stocks

# Call the function with your three dataframes
darkcloud_stocks = detect_darkcloud(df, df1, df2,df3)

print("Stocks in which darkcloud candlesticks were formed:")
for stock in darkcloud_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which darkcloud candlesticks were formed': darkcloud_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_darkcloud.xlsx", index=False)

def detect_evngstar(df, df1, df2,df3):
    evngstar_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for Hanging Man pattern
        if (row2_prev['CLOSE'] > row2_prev['OPEN']) and \
           (row3_prev['CLOSE'] > row3_prev['OPEN']) and \
           (row2_prev['CLOSE'] > row3_prev['CLOSE']) and \
           abs((row2_prev['OPEN'] - row2_prev['CLOSE'])) > 0.1*(row2_prev['PREVCLOSE']) and \
           (row1_prev['OPEN'] > row1_prev['PREVCLOSE']) and \
           (abs(row1_prev['OPEN']-row1_prev['CLOSE']) < 0.05*(row1_prev['HIGH']-row1_prev['LOW'])) and\
           row['OPEN'] > row['CLOSE'] and \
           row['OPEN'] - row['CLOSE'] > 0.75 * (abs(row2_prev['OPEN'] - row2_prev['CLOSE'])):
                evngstar_stocks.append(symbol)

    return evngstar_stocks

# Call the function with your three dataframes
evngstar_stocks = detect_evngstar(df, df1, df2,df3)

print("Stocks in which evngstar candlesticks were formed:")
for stock in evngstar_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which evngstar candlesticks were formed': evngstar_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_evngstar.xlsx", index=False)

def detect_shootingstar(df, df1, df2,df3):
    shootingstar_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for Hanging Man pattern
        if (row1_prev['CLOSE'] > row1_prev['OPEN']) and \
           (row2_prev['CLOSE'] > row2_prev['OPEN']) and \
           (row3_prev['CLOSE'] > row3_prev['OPEN']) and \
           (row1_prev['CLOSE'] > row3_prev['CLOSE']):
            if 2 * abs(row['OPEN'] - row['CLOSE']) < (row['HIGH'] - max(row['OPEN'], row['CLOSE'])) and \
               min(row['OPEN'], row['CLOSE']) - row['LOW'] < 0.05 * (row['HIGH'] - row['LOW']):
                shootingstar_stocks.append(symbol)

    return shootingstar_stocks

# Call the function with your three dataframes
shootingstar_stocks = detect_shootingstar(df, df1, df2,df3)

print("Stocks in which Shootingstar candlesticks were formed:")
for stock in shootingstar_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which shooting candlesticks were formed': shootingstar_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_shootingstar.xlsx", index=False)

def detect_bearish_harami(df, df1, df2,df3):
    bearish_harami_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for bullish engulfing pattern
        if (row1_prev['CLOSE'] > row1_prev['OPEN']) and \
           (row2_prev['CLOSE'] > row2_prev['OPEN']) and \
           (row3_prev['CLOSE'] > row3_prev['OPEN']) and \
           (row1_prev['CLOSE'] > row3_prev['CLOSE']) and \
           (row['CLOSE'] < row['OPEN']) and\
           (row['OPEN'] < row1_prev['CLOSE']) and \
           (row['CLOSE'] > row1_prev['OPEN']):
                bearish_harami_stocks.append(symbol)

    return bearish_harami_stocks

# Call the function with your three dataframes
bearish_harami_stocks = detect_bearish_harami(df, df1, df2,df3)

print("Stocks in which bearish harami candlesticks were formed:")
for stock in bearish_harami_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which bearish harami candlesticks were formed': bearish_harami_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_bearishharami.xlsx", index=False)

def detect_rmarubozu(df):
    rmarubozu_stocks = []

    for index, row in df.iterrows():
        if abs(row['CLOSE']-row['OPEN']) > 0.1*(row['PREVCLOSE']) and\
           row['CLOSE'] < row['OPEN'] and\
           row['HIGH']-row['OPEN'] < 0.01*(row['HIGH']-row['LOW']) and\
           row['CLOSE']-row['LOW'] < 0.01*(row['HIGH']-row['LOW']):
                rmarubozu_stocks.append(row['SYMBOL'])
    return rmarubozu_stocks

# Detect Doji candlesticks for stocks
rmarubozu_stocks = detect_rmarubozu(df)

print("Stocks in which rmarubozu candlesticks were formed:")
for stock in rmarubozu_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which red marubozu candlesticks were formed': rmarubozu_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_rmarubozu.xlsx", index=False)

def detect_blackcrows(df, df1, df2,df3):
    blackcrows_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for bullish engulfing pattern
        if (row1_prev['CLOSE'] < row1_prev['OPEN']) and \
           (row2_prev['CLOSE'] < row2_prev['OPEN']) and \
           (row1_prev['CLOSE'] < row2_prev['CLOSE']) and \
           (row['CLOSE'] < row['OPEN']) and\
           (row2_prev['CLOSE']< min(row3_prev['OPEN'],row3_prev['CLOSE'])) and\
           (row1_prev['OPEN']-row1_prev['CLOSE']) > 0.1*(row1_prev['PREVCLOSE']) and\
           (row['OPEN']-row['CLOSE']) > 0.1*(row['PREVCLOSE']):
                blackcrows_stocks.append(symbol)

    return blackcrows_stocks

# Call the function with your three dataframes
blackcrows_stocks = detect_blackcrows(df, df1, df2,df3)

print("Stocks in which blackcrows candlesticks were formed:")
for stock in blackcrows_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which blackcrows candlesticks were formed': blackcrows_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_blackcrows.xlsx", index=False)

def detect_tweezertop(df, df1):
    tweezertop_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for Hanging Man pattern
        if (row1_prev['CLOSE'] > row1_prev['OPEN']) and \
           (row['CLOSE'] < row['OPEN']) and \
           abs(row['OPEN'] - row['PREVCLOSE']) < 0.01*(row['HIGH']-row['LOW']):
                tweezertop_stocks.append(symbol)

    return tweezertop_stocks

# Call the function with your three dataframes
tweezertop_stocks = detect_tweezertop(df, df1)

print("Stocks in which tweezer top candlesticks were formed:")
for stock in tweezertop_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which tweezertop candlesticks were formed': tweezertop_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_tweezertop.xlsx", index=False)

def detect_bearishca(df, df1, df2,df3):
    bearishca_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for bullish engulfing pattern
        if (row1_prev['CLOSE'] > row1_prev['OPEN']) and \
           (row2_prev['CLOSE'] > row2_prev['OPEN']) and \
           (row3_prev['CLOSE'] > row3_prev['OPEN']) and \
           (row1_prev['CLOSE'] > row3_prev['CLOSE']) and \
           (row['CLOSE'] < row['OPEN']) and\
           (row['OPEN'] > row['PREVCLOSE']) and\
           abs(row['CLOSE']-row['PREVCLOSE']) < 0.01*(row['HIGH']-row['LOW'])  :
                bearishca_stocks.append(symbol)

    return bearishca_stocks

# Call the function with your three dataframes
bearishca_stocks = detect_bearishca(df, df1, df2,df3)

print("Stocks in which bearish counterattack candlesticks were formed:")
for stock in bearishca_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which bearish counterattack candlesticks were formed': bearishca_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_bearishcounterattack.xlsx", index=False)

def detect_dt(df, df1, df2,df3):
    dt_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for bullish engulfing pattern
        if (row1_prev['CLOSE'] < row1_prev['OPEN']) and \
           (row['CLOSE'] < row['OPEN']) :
                dt_stocks.append(symbol)

    return dt_stocks

# Call the function with your three dataframes
dt_stocks = detect_dt(df, df1, df2,df3)

print("Stocks in which dt were formed:")
for stock in dt_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which dt candlesticks were formed': dt_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_dt.xlsx", index=False)

def detect_gd(df, df1, df2,df3):
    gd_stocks = []

    for index, row in df.iterrows():
        symbol = row['SYMBOL']
        
        # Get corresponding rows from df1 and df2
        row1_prev = df1[df1['SYMBOL'] == symbol].iloc[0]
        row2_prev = df2[df2['SYMBOL'] == symbol].iloc[0]
        row3_prev = df3[df3['SYMBOL'] == symbol].iloc[0]

        # Conditions to check for bullish engulfing pattern
        if (min(row1_prev['OPEN'],row1_prev['CLOSE'])-row['OPEN']) > 0.015*(min(row1_prev['OPEN'],row1_prev['CLOSE'])):
                gd_stocks.append(symbol)

    return gd_stocks

# Call the function with your three dataframes
gd_stocks = detect_gd(df, df1, df2,df3)

print("Stocks in which gd were formed:")
for stock in gd_stocks:
    print(stock)

# Create a DataFrame from the list of stocks
output_df = pd.DataFrame({'Stocks in which gd candlesticks were formed': gd_stocks})

# Export the DataFrame to an Excel file
output_df.to_excel("output_gd.xlsx", index=False)
```
